## Lecture 2: Normal Curve

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (10.0, 6.0)
pd.options.display.float_format = '{:,.2f}'.format
pd.set_option('max_columns', 50)
pd.set_option('max_rows', 9)
```

### Standard Units

```python
def std_units(nums):
    ''' convery array to standard units '''
    return ((nums - np.mean(nums)) / np.std(nums))
```

```python
births = pd.read_csv('data/baby.csv')
age_std_units = std_units(births['Maternal Age'])
print(np.mean(age_std_units), np.std(age_std_units))


age_plus_std_units = pd.DataFrame({'Age': births['Maternal Age'],
                                   'Age std units': age_std_units})
age_plus_std_units[:9]
```

```python
births['Maternal Age'].hist(bins=np.arange(15, 46, 2), ec='w')
plt.show()
```

```python
avg = np.mean(births['Maternal Age'])
std = np.std(births['Maternal Age'])
bins = np.arange((15 - avg) / std, (46 - avg) / std, 2  / std)

age_plus_std_units['Age std units'].hist(bins=bins, ec='w')
plt.show()
```

### SD and Bell Curves

On a (roughly) bell shaped curve, +/- 1 standard deviation occurs at the points of inflection

```python
births['Maternal Height'].hist(bins=np.arange(56.5, 72.6, 1), ec='w')
plt.show()

print(np.mean(births['Maternal Height']))
print(np.std(births['Maternal Height']))
```

### The central limit theorem

```python
united = pd.read_csv('data/united_summer2015.csv')
united['Delay'].hist(bins=np.arange(-20, 300, 10), ec='w', density=True)
plt.show()
```

```python
sample_size = 500
averages = np.array([np.mean(united['Delay'].sample(n=sample_size, replace=True))
            for _ in range(10000)])

plt.hist(averages, bins=25, ec='w')
plt.show()

print(np.mean(united['Delay']), np.std(united['Delay']) / np.sqrt(sample_size))
```
