#### Hypothesis testing and Confidence intervals

```python
def cohen_d(x, y):
    ''' Returns Cohen's d for two data sets'''
    n_x = len(x)
    x_bar = np.mean(x)
    s_x = np.std(x, ddof=1)
    n_y = len(y)
    y_bar = np.mean(y)
    s_y = np.std(y, ddof=1)

    s_pooled = np.sqrt((((n_x - 1) * s_x ** 2) +
                        ((n_y - 1) * s_y ** 2)) / (n_x + n_y - 2))
    cohen_d = np.abs(x_bar - y_bar) / s_pooled

    return cohen_d
```

```python
def ci_diff(x, y, cl=0.95, eqvar=False):
    ''' Returns the confidence interval for the difference of two data sets '''
    n_x = len(x)
    x_bar = np.mean(x)
    s_x = np.std(x, ddof=1)
    n_y = len(y)
    y_bar = np.mean(y)
    s_y = np.std(y, ddof=1)
    
    # Difference in means
    diff_bar = x_bar - y_bar
  
    # dof - if population variences are equal
    if eqvar:
        dof = n_x + n_y - 2
        std_diff = np.sqrt(((n_x - 1) * (s_x) ** 2 +
                            (n_y - 1) * (s_y) ** 2) / 
                           dof * ((1 / n_x) + (1 / n_y)))
        
    else:
        dof = ((s_x ** 2 / n_x + s_y ** 2 / n_y) ** 2 / 
               (1 / (n_x - 1) * (s_x ** 2 / n_x) ** 2 + 
                1 / (n_y - 1) * (s_y ** 2 / n_y) ** 2))
        std_diff = np.sqrt(s_x ** 2 / (n_x) + s_y ** 2 / (n_y))
        
    # t-score
    alpha = 1 - cl
    t_score = (stats.t.ppf((1 - (alpha / 2)), dof)) 
    
    # Error bound of the mean  
    ebm = t_score * std_diff 
    

    return (diff_bar - ebm, diff_bar + ebm)
```

```python
for col in baby.drop('Maternal Smoker', axis=1).columns:
    smoker = baby.loc[baby['Maternal Smoker'] == True][col]
    nonsmoker = baby.loc[baby['Maternal Smoker'] == False][col]
    print('{0}:'.format(col))
    print(' p-value   = {0:.3f}'.format(stats.ttest_ind(smoker, nonsmoker, 
                                               equal_var=False)[1]))
    print(" Cohen's d = {0:.3f}".format(cohen_d(smoker, nonsmoker)))
    ci_low, ci_high = ci_diff(smoker, nonsmoker)
    print(" 95% CI of difference (smoker - nonsmoker) = ({0:.2f}, {1:.2f})".
          format(ci_low, ci_high))
    print()
```

#### Using t-tests

```python
# Blakeman data
pats_blakeman_data = (deflate_data.loc[deflate_data['Team'] == 'Patriots']
                                           ['Blakeman'].values)

colts_blakeman_data = (deflate_data.loc[deflate_data['Team'] == 'Colts']
                                           ['Blakeman'].values)
```

```python
stats.ttest_ind(pats_blakeman_data, colts_blakeman_data)
```

```python
# Prioleau data
pats_prioleau_data = (deflate_data.loc[deflate_data['Team'] == 'Patriots']
                                           ['Prioleau'].values)

colts_prioleau_data = (deflate_data.loc[deflate_data['Team'] == 'Colts']
                                           ['Prioleau'].values)
```

```python
stats.ttest_ind(pats_prioleau_data, colts_prioleau_data)
```

#### Using simulation

```python
deflate_data['Avg'] = (deflate_data['Blakeman'] + deflate_data['Prioleau']) / 2

pats_start = 12.5
colts_start = 13.0
deflate_data['Drop'] = deflate_data.apply(lambda srs: pats_start - srs['Avg']
                                          if srs['Team'] == 'Patriots'
                                          else colts_start - srs['Avg']
                                          , axis=1)
```

```python
actual_drops = deflate_data.groupby(['Team'])['Drop'].agg('mean')

display(actual_drops)

obs_diff = actual_drops[1] - actual_drops[0]
obs_diff
```

```python
sim_drops_diff = []
drops = deflate_data['Drop'].values

for _ in range(10000):
    np.random.shuffle(drops)
    sim_drops_diff.append(drops[:11].mean() - drops[11:].mean())
```

```python
plt.hist(sim_drops_diff, density=True, bins=30)
plt.scatter(obs_diff, 0, color='red', s=60, zorder=2);
plt.show()
```
